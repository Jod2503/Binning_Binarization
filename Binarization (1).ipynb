```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df= pd.read_csv('train.csv',usecols=['Age','Fare','SibSp','Parch','Survived'])
```

```python
df['Family']=df['SibSp']+df['Parch']
```

```python
df
```

```python
df.drop(columns=['SibSp','Parch'])
```

```python
df.dropna(inplace=True)
```

```python
# with binarization
x_train,x_test,y_train,y_test=train_test_split(df[['Age','Fare','Family']],df['Survived'],test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import Binarizer
```

```python
# we need to find if person is travelling with a family or not
from sklearn.compose import ColumnTransformer
```

```python
trf=ColumnTransformer(transformers=[
    ('trf1',Binarizer(copy=False),['Family'])],
                       remainder='passthrough')                    
```

```python
x_train_tf=trf.fit_transform(x_train)
x_test_tf=trf.transform(x_test)
```

```python
x_train_tf
```

```python
x_train_tf=pd.DataFrame(x_train_tf,columns=['family','Age','Fare'])
```

```python
x_train_tf
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(x_train_tf,y_train)
y_pred=clf.predict(x_test_tf)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
